Machine learning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
y = df['default payment next month']
X = df.drop(columns = 'default payment next month')
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
X_train
y_train
```

```python
num_cols = X.select_dtypes('number').columns
cat_columns= X.select_dtypes('object').columns
print(f'Numerical columns: {num_cols}\nCategorical columns: {cat_columns}')
```

```python
numerical_imputer = SimpleImputer(strategy='mean')
scaler = StandardScaler()
numerical_transformer = Pipeline(steps=[
    ('num_imputer', numerical_imputer),
    ('scaler', scaler)
])
```

Logistic Regression Method


```python
default_pipeline = Pipeline(steps=[
    ('preprocessor', numerical_transformer),
    ('classifier', LogisticRegression())
])
```

```python
default_pipeline
```

```python
#Training phase
default_pipeline.fit(X_train, y_train)
```

```python
y_pred = default_pipeline.predict(X_test)
y_pred
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap= 'Reds')
plt.title('Performance of the LR model in predicting default payment next month');
plt.show()
```

Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier

RANDOM_STATE = 42

# Scale the data using the defined numerical_transformer
X_train_scaled = numerical_transformer.fit_transform(X_train)
X_test_scaled = numerical_transformer.transform(X_test)

#Random Forest Classfication
rf = RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, class_weight='balanced')
rf.fit(X_train_scaled, y_train)
y_pred_rf = rf.predict(X_test_scaled)

print("Random Forest accuracy:", accuracy_score(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_rf, cmap= 'Blues')
plt.title('Performance of the RF model in predicting default payment next month');
plt.show()
```

Neural Networks model

```python

import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout


# Define EPOCHS and BATCH_SIZE
EPOCHS = 50
BATCH_SIZE = 32

n_classes = len(np.unique(y))
if n_classes > 2:
    final_activation = 'softmax'
    loss = 'sparse_categorical_crossentropy'
else:
    final_activation = 'sigmoid'
    loss = 'binary_crossentropy'

model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dense(n_classes if n_classes>2 else 1, activation=final_activation)
])

model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])
history = model.fit(X_train_scaled, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, validation_split=0.15, verbose=1)
```

```python
plt.figure(figsize=(12, 5))

# Plot training & validation accuracy values
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.tight_layout()
plt.show()
```

```python
#Evaluate model
y_pred_prob = model.predict(X_test_scaled)
if n_classes > 2:
    y_pred = y_pred_prob.argmax(axis=1)
else:
    y_pred = (y_pred_prob.ravel() > 0.5).astype(int)

print("Neural network accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
